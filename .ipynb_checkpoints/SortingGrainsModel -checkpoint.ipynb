```python
import numpy as np 
import os
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from skimage.io import imread, imshow
```

```python
paths_bad = os.listdir('../Pict1/')
paths_good = os.listdir('../Pict2/')
all_img = []
y_train = []
for img in paths_bad:
    y_train.append(0)
    all_img.append(os.path.join('../Pict1/', img))
    
for img in paths_good:
    y_train.append(1)
    all_img.append(os.path.join('../Pict2/', img))
```

```python
from tqdm import tqdm_notebook
```

```python
list_of_image = []

for name in tqdm_notebook(all_img):
    img = imread(name)
    full_img = np.zeros([60, 60, 3])
    h, w = img.shape[:2]
    full_img[:h, :w, :] = img
    list_of_image.append(full_img)

img_array = np.zeros([len(list_of_image), 60, 60, 3])

for i in range(len(list_of_image)):
    img_array[i] = list_of_image[i]
```

```python
plt.figure(figsize=(12, 7))
height, width = 3, 5
for i in range(1, height * width + 1):
    plt.subplot(height, width, i)
    plt.title(y_train[i])
    plt.imshow(img_array[i])
```

```python
img_array.shape
```

```python
from skimage.feature import hog
```

```python
img_array_grayScale = np.zeros((len(img_array), 60, 60))
```

```python
from skimage.color import rgb2gray
```

```python
for i in range(len(img_array)):
    img = img_array[i]
    img_gray = rgb2gray(img)
    img_array_grayScale[i] = img_gray
```

```python
img_array_grayScale.shape
```

## Computing HoGs

```python
X_train = np.zeros((len(img_array_grayScale), 2025))
```

```python
for i in tqdm_notebook(range(len(img_array_grayScale))):
    img = img_array_grayScale[i]
    curr_hog = hog(img)
    X_train[i] = curr_hog
```

```python
X_train.shape
```

```python
len(y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, X_test, y, y_test = train_test_split(X_train, y_train
                                        , test_size=0.33
                                        , random_state = 42
                                        , shuffle=True)
```

```python
knn = KNeighborsClassifier(n_neighbors=15, n_jobs=-1)
```

```python
knn.fit(X, y)
```

```python
ans = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, ans)
```
